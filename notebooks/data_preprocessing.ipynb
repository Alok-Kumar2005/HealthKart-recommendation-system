```python
## Data only required for the review analysis is review.text and review.rating
df = df[["reviews.text", "reviews.rating"]]
```

```python
## function to get the number of words in a text
def word_count(text):
    return len(str(text).split())

df['word_count'] = df['reviews.text'].apply(word_count)
df.head()
```

```python
print("Average word count:", df['word_count'].mean())
print("Min word count:", df['word_count'].min())
print("Max word count:", df['word_count'].max())
```

- Dataset contain more than 70000 rows, so not possible to preprocess, feature engineering and model training in my system
- I am taking only 40000 rows for training

```python
## taking only 40000 samples for analysis
df = df.sample(n=40000, random_state=42)
df = df.reset_index(drop=True)  
df.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
def transform_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word.isalnum()]
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [ps.stem(word) for word in tokens]

    return " ".join(tokens)
```

```python
df['transformed_text'] = df['reviews.text'].apply(transform_text)
```
