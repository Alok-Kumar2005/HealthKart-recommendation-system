```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
result = classifier("I love using transformers library!")
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
sentiment = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
result = sentiment("This phone is amazing")
print(result)
```

```python
result = sentiment("This phone is good but could be better")
print(result)
```

```python
result = classifier("This phone is good but could be better and also i will not recommend it to anyone but at that price it is nice.")
print(result)
```
